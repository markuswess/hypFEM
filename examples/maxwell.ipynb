## Derivation of the hyperbolic compactified formulation

We start from the time-domain Maxwell system in weak first-order $E-H$ formulation

$$
\partial_t \int_\Omega \tilde E\cdot \tilde e =\int_\Omega\nabla\times \tilde H\cdot  \tilde e,\\
\partial_t \int_\Omega \tilde H\cdot \tilde h = -\int_\Omega \nabla\times \tilde E\cdot  \tilde h
$$
for all test functions $e,h$, where $\Omega$ is some suitable domain.



Introducing the time transformation $\nabla = -K\partial_\tau+\nabla$ (random choice for the usual $H$ to be replaced by $K$ for obvious reasons...) we obtain
$$
\partial_\tau \int_\Omega E\cdot  e =\int_\Omega(-K\partial_\tau+\nabla)\times H\cdot  e=-\int_\Omega K\times\partial_\tau H\cdot  e+\int_\Omega \nabla\times H\cdot  e,\\
\partial_\tau \int_\Omega H\cdot  h =-\int_\Omega(-K\partial_\tau+\nabla)\times E\cdot  h =\int_\Omega K\times\partial_\tau E\cdot h-\int_\Omega \nabla\times E\cdot  h
$$
Replacing $\partial_\tau E, \partial_\tau H$ and moving terms we obtain


$$
\partial_\tau \int_\Omega (E+K\times K\times E)\cdot  e =\int_\Omega K\times\nabla\times E\cdot  e+\int_\Omega \nabla\times H\cdot  e,\\
\partial_\tau \int_\Omega (H+K\times K \times H)\cdot  h =\int_\Omega K\times\nabla\times H\cdot h-\int_\Omega \nabla\times E\cdot  h
$$

Next we apply a spacial Compactification with Jacobian $J$ and use covariant transformations $E\mapsto J^{-T} E$, $H\mapsto J^{-T}H$ for the fields to obtain
$$
\partial_\tau \int_\Omega (J^{-T}E+K\times K\times J^{-T}E)\cdot  J^{-T}e\det J =\int_\Omega K\times J\nabla\times E\cdot  J^{-T}e+\int_\Omega \nabla\times H\cdot  e,\\
\partial_\tau \int_\Omega (J^{-T}H+K\times K \times J^{-T}H)\cdot  J^{-T}h\det J =\int_\Omega K\times J\nabla\times H\cdot J^{-T}h-\int_\Omega \nabla\times E\cdot  h
$$

Using a radial scaling we have
$$
J=r'\Pi_x+\frac{r}{\rho}\Pi_x^\perp, \qquad \det J = \frac{r'r^2}{\rho^2}
$$
with
$$
\begin{align}
\Pi_{x} &= \frac{ xx^T}{\|x\|^2},&\Pi_x^\perp = I-\frac{xx^T}{\|x\|^2}
\end{align}
$$
and also 
$$
K = \frac{k}{\rho}x.
$$


Thus, we find (sign?)
$$
(K\times J\nabla\times E)\cdot J^{-T}e = (\frac{kr}{\rho}\Pi_x^\perp(\nabla\times E))(\frac{\rho}{r}\Pi_x^\perp+\frac{1}{r'}\Pi_x) e=k\Pi_x^\perp(\nabla\times E)\cdot\Pi_x^\perp e
$$
and
$$
\begin{align}
(J^{-T}E+K\times K\times J^{-T}E)\cdot  J^{-T}e\det J &= (J^{-T}E -k^2\Pi_x^\perp(J^{-T}E))\cdot J^{-T}e\det J\\
&=\frac{1}{r'}\Pi_x E\cdot e+(1-k^2)r'\Pi^\perp_x E\cdot e
\end{align}
$$
and similar for the $H$ terms

### Anil derivation

Using your notation
$$
\Pi_x=nn^T,\qquad \Pi_x^\perp=I-\Pi_x,\\

J=r'\Pi_x+\frac r\rho \Pi_x^\perp,\\
J^{-T}=\frac1{r'}\Pi_x+\frac{\rho}{r}\Pi_x^\perp,\\
\det J=r'\frac{r^2}{\rho^2}, \qquad K=kn.
$$


Identities are
$$
(K\times J\nabla_x\times E)\cdot J^{-T}e
=
k\bigl(n\times(\nabla_x\times E)\bigr)\cdot e\\

\begin{aligned}
&\left(
J^{-T}E+K\times K\times J^{-T}E
\right)\cdot J^{-T}e\,\det J
\\
&\qquad =
\frac{r^2}{r'\rho^2}\,\Pi_x E\cdot e
+
(1-k^2)r'\,\Pi_x^\perp E\cdot e .
\end{aligned}
$$

Compactified mass tensor
$$
\mathsf M =
\frac{r^2}{r'\rho^2}\,\Pi_x +
(1-k^2)r'\,\Pi_x^\perp = \frac1L\,\Pi_x + \frac{1-k^2}{G}\,\Pi_x^\perp .
$$

The compactified Maxwell system
$$
\mathsf M\,\partial_\tau E = \nabla_x\times H + k\,n\times(\nabla_x\times E),\\
\mathsf M\,\partial_\tau H = k\,n\times(\nabla_x\times H) - \nabla_x\times E.
$$

Final equations
$$
\left(
\frac1L\,\Pi_x
+
\frac{1-k^2}{G}\,\Pi_x^\perp
\right)\partial_\tau E
=
\nabla_x\times H +k\,n\times(\nabla_x\times E),\\
\frac1L\,\Pi_x+ \frac{1-k^2}{G}\,\Pi_x^\perp)\partial_\tau H
=
k\,n\times(\nabla_x\times H)
-
\nabla_x\times E.
$$
